```python
#homework2 problem3
import numpy as np
Genome=""
k=11
L=566
t=18
with open("rosalind_ba1e.txt",'r') as f:
    Genome=f.readline().strip()
    k,L,t=f.readline().strip().split()

k=int(k)
L=int(L)
t=int(t)
frqtable=np.zeros(4**k) #pattern frequency table
subGenome=Genome[0:L]

def NtoD(kmer):#nuclueotide to decimal
    dec=0
    k=len(kmer)
    d={'A':0,'C':1,'G':2,'T':3}
    for j in range(0,k):
        dec += d[kmer[j]]*(4**(k-j-1))
    return dec

for i in range(0,L-k):
    kmer=subGenome[i:i+k]
    frqtable[NtoD(kmer)]+=1
    
def DtoQ(dec): #decimal to quaternary
    result = ''
    while dec>0:
        qua = dec % 4
        dec = dec // 4
        result=str(qua)+result
    return int(result)

def QtoN(qua,k): #quaternary to nucleotide
    d={'0':'A','1':'C','2':'G','3':'T'}
    result=''
    s=str(qua)
    for a in s:
        result+=d[a]
    if k>len(s):  # if there are zeros before the first non-zero, and k-len(s) number of A
        for i in range(0,k-len(s)):
            result='A'+result
    return result
indices=np.where(frqtable>=t)
indiceslist=[]
for i in range(0,len(indices[0])):
    indiceslist.append(QtoN(DtoQ(indices[0][i]),k))


for i in range(1,len(Genome)-L+1):
    newSubGenome=Genome[0+i:L+i]
    oldSubGenome=Genome[0+i-1:L+i-1]
    lostkmer=oldSubGenome[0:k]
    gainkmer=newSubGenome[L-k:L]

    frqtable[NtoD(lostkmer)]-=1
    frqtable[NtoD(gainkmer)]+=1

    if frqtable[NtoD(gainkmer)] >=t:
        if gainkmer not in indiceslist:
            indiceslist.append(gainkmer)
print(*indiceslist)
```

```python
#problem 16
import numpy as np
frqtable={}

def rvcmp(s): #get reverse complementary sequence
    sc=""
    d={'A':'T','T':'A','G':'C','C':'G'}
    for a in s:
        sc=d[a]+sc
    return sc

def dH(s,t):
    d=0 #initialize distance =0
    for i in range(0,len(s)):
        if s[i] != t[i]:
            d+=1
    return d

with open("rosalind_corr.txt",'r') as f:
    for line in f:
        if line[0]!='>':
            s=line.strip()
            if s in frqtable:
                frqtable[s]+=1
            elif rvcmp(s) in frqtable:
                frqtable[rvcmp(s)]+=1
            else:
                frqtable[s]=1
correctlist=[]
putativewrong={}
for key,value in frqtable.items():
    if value >=2:
        correctlist.append(key)
    else:
        putativewrong[key]=0
for i in putativewrong:
    match=''
    for j in correctlist:
        if dH(i,j)==1 :
            putativewrong[i]+=1
            match=j
        elif dH(i,rvcmp(j))==1:
            putativewrong[i]+=1
            match=rvcmp(j)
    if putativewrong[i]==1:
        print(i+'->'+match)

```
