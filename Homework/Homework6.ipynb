```python
#problem 39
d={}

with open('RNAcodetable.txt','r') as f:
    for line in f:
        temp=line.strip().split()
        for key,value in zip(temp[0::2],temp[1::2]):
            d[key]=value
with open('rosalind_prot.txt','r') as f:
    for line in f:
        mRNA=line.strip()
protein=''
startpos=0
for i in range(len(mRNA)-3):
    if mRNA[i:i+3]=='AUG':
        protein+='M'
        startpos=i
        break

for i in range(startpos+3,len(mRNA)-3,3):
    if d[mRNA[i:i+3]]!='Stop':
        protein+=d[mRNA[i:i+3]]  
    else:
        break

protein
            
```

```python
#problem 40
L=[]
with open('RNAcodetable.txt','r') as f:
    for line in f:
        temp=line.strip().split()
        for key in temp[1::2]:
            L.append(key)
import collections as cn
c=cn.Counter(L)
protein=''
with open('rosalind_mrna.txt','r') as f:
    protein=f.readline().strip()
res=1
for aa in protein:
    res=res*c[aa]
    if res > 1000000:
        res = res % 1000000

# for stop codon only
res=res*3 
if res > 1000000:
    res = res % 1000000
res
```
