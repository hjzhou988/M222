```python
#DFS, test whether a graph is connected or not
def dfs(graph, start):
    visited, stack = set(), [start]
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            visited.add(vertex)
            stack.extend(graph[vertex] - visited)
    if len(graph)==len(visited):
        return True
    else:
        return False
dfs(d,'AA')
```
