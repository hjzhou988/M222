```python
#problem 28
import numpy as np
s=""
StringList=[]
with open("rosalind_pdst.txt",'r') as f:
    next(f)
    for line in f:
        if line[0]=='>':
            StringList.append(s)
            s=''#start an empty string
        else:
            s=s+line.strip()
    StringList.append(s) #string of the last sequence
def pD(s,t):
    L=len(s)
    mismatch=0
    for i in range (L):
        if s[i]!=t[i]:
            mismatch+=1
    return mismatch/L

N=len(StringList)    
D=np.zeros((N,N))  
for i in range(N):
    for j in range(i+1,N):
        D[j,i]=D[i,j]=pD(StringList[i],StringList[j])
np.set_printoptions(precision=5)
print(D)
        
```

```python
with open("rosalind_hdag.txt",'r') as f:
    Total=int(f.readline().strip())
    for i in range(Total):
        #tempty=f.readline()# There is an empty line between each graph
        nodes,edges=f.readline().strip().split()
        nodes=int(nodes)
        edges=int(edges)
        dic={}
        Indeg={}
        for j in range(1,nodes+1): #initialize the Indeg and Outdeg dictionary
            Indeg[j]=0
        for j in range(edges):
            a,b=f.readline().strip().split()
            a=int(a)
            b=int(b)
            if a in dic:
                dic[a].append(b)
            else:
                dic[a]=[b]
            Indeg[b]+=1
        
        Path=[]
        containsHPath=True
        while len(Indeg.keys())>0:
            startnode=-1
            count=0
            for key,value in Indeg.items():
                if value==0:
                    startnode=key
                    count+=1  
            if count>1:
                containsHPath=False
                break
            else:
                Indeg.pop(startnode)
                Path.append(startnode)
                if startnode not in dic:
                    break
                for nodes in dic[startnode]:
                    Indeg[nodes]-=1
        if containsHPath==False:
            print(-1)
        else:
            print(*([1]+Path))
            
```
